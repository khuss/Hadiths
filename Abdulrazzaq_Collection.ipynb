```python
def is_numb(word):
    for w in word:
        if w  in ar_num:
            isit = True
        else:
            isit = False
            break
    return isit

def cut(words):
    sep = 0
    for w in range(0,len(words)-3):
        w1 = words[w]
        w2 = words[w+1]
        w3 = words[w+2]
        b1 = is_numb(w1)
        b2 = is_numb(w2)
        b3 = is_numb(w3)
        if b1 and b2 and b3 :
            sep = w
            break
    return words[sep:]
                    
        
def translate(num):
    try:
        eng_equi=''
        for i in range(0,len(num)):
                temp = ar_num.index(num[i])
                eng_equi += eng_num[temp]
    except:
        temp = 0
    return eng_equi

def switch(words):
    ind = 0
    if (words[0] == 'أَخْبَرَنَا'):
        #print('check')
        for i in range(len(words)-1):
            w1 = is_numb(words[i])
            w2 = words[i+1]
            if w1 and (w2 == '-'):
                words[0],words[ind] = words[ind], words[0]
                words[1],words[ind+1]=words[ind+1], words[1]
                break
    elif(words[0:2] == ['عَبْدُ' ,'الرَّزَّاقِ،']):
        for i in range(len(words)-1):
            w1 = is_numb(words[i])
            w2 = words[i+1]
            if w1 and (w2 == '-'):
                words[0],words[ind] = words[ind], words[0]
                words[1],words[ind+1]=words[ind+1], words[1]
                break
        
        

def multiplicity(worded_hadith):
    ind = []
    for i in range(len(worded_hadith)-1):
        b1 = is_numb(worded_hadith[i])  
        b2 = (worded_hadith[i+1])
        if (b1 and (b2 == '-')):
            ind.append(i)
  #  if len(ind) == 0:
  #      ind = [0]
    return ind
            


hadith = []  
index = []
hadith_number = []
ar_num = ['٨','٦','٩','٥','٣','٧','٢','٤','١',' ','-','(', ')','م', '٠','ب']
eng_num = ['8', '6', '9', '5', '3', '7', '2', '4','1', ' ', '-', '(', ')', 'M','0','B']
count = 0
for i in df_trial:
    j = " ".join(i)
    w = j.split()
    switch(w)
    ind = multiplicity(w)
    if len(w)>10:
        for k in range(len(ind)):
            if is_numb(w[ind[k]]):
                num = w[ind[k]]
            else:
                num = 0
            if k!= len(ind)-1:
                had = w[ind[k]:ind[k+1]]
                hadith.append(" ".join(had))
                hadith_number.append(translate(num))
                index.append(df_trial.index[count])
            else:
                hadith.append(" ".join(w[ind[k]:len(w)]))
                hadith_number.append(translate(num))
                index.append(df_trial.index[count]) 
    count += 1
        
```
