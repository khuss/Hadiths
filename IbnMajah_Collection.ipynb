```python
import json
import pandas as pd
import numpy as np
```

```python
a= open ('dump.json', "r")
data=json.load(a)
df = pd.read_json("dump.json")
```

```python
df_ibnmajah = df.loc[df['book_id']==1198]
```

```python
df_trial = df_ibnmajah.loc[:,'clean_texts']
```

```python
def is_numb(word):
    for w in word:
        if w  in ar_num:
            isit = True
        else:
            isit = False
            break
    return isit


        
def translate(num):
    try:
        eng_equi=''
        for i in range(0,len(num)):
                temp = ar_num.index(num[i])
                eng_equi += eng_num[temp]
    except:
        temp = 0
    return eng_equi

    
hadith = []  
index = []
hadith_number = []
ar_num = ['٨','٦','٩','٥','٣','٧','٢','٤','١',' ','-','(', ')','م', '٠','ب']
eng_num = ['8', '6', '9', '5', '3', '7', '2', '4','1', ' ', '-', '(', ')', 'M','0','B']
count = 0
for i in df_trial:
    j = " ".join(i)
    w = j.split()
    if len(j)>200:
        hadith.append(j)
        index.append(df_trial.index[count])
        if(is_numb(w[0])):
            hadith_number.append(translate(w[0]))
        else:
            hadith_number.append(0)
    else:
        if(is_numb(w[0])):
            hadith.append(j)
            index.append(df_trial.index[count])
            hadith_number.append(translate(w[0]))
            
        
    count+=1
```

```python
len(hadith)
```

```python
new_df = pd.DataFrame(hadith,index = index,columns=['clean_texts'])
hadith_number = pd.DataFrame(hadith_number, columns = ['hadith_number'])
```

```python
new_df.drop(columns=['clean_texts'], inplace = True)
new_df = new_df.join(df_ibnmajah,how='left')
new_df = new_df.reset_index()
```

```python
new_df = pd.concat([new_df, hadith_number],axis=1)
```

```python
sorted_df = new_df.loc[:,['shamela_link', 'page_id', 'clean_texts', 'hadith_number']].sort_values(by = 'page_id')
```
