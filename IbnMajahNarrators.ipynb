```python
import json
import pandas as pd
import numpy as np
```

```python
a= open ('Final_IbnMajah', "r")
data=json.load(a)
df = pd.read_json("Final_IbnMajah")
```

```python
def read_hadiths(file):
    a= open (file, "r")
    data=json.load(a)
    df = pd.read_json(file)
    df = df.loc[:,'clean_texts']
    sentence = []
    label = []
    data = []
    for i in df:
        j = " ".join(i)
        w = j.split()
        for g in w:
            sentence.append(g)
            label.append('O')
        data.append(sentence)
        sentence = []
        label = []
    return data
```

```python
hadith = np.array(read_hadiths('Final_IbnMajah'))
hadith = hadith[:,0].tolist()
```

```python
ibn_majah_predictions = pd.read_csv('ibn_m_pred')
```

```python
ibn_majah_predictions = ibn_majah_predictions.iloc[0]
```

```python
data = []
for i in range(1,len(ibn_majah_predictions)):
    isnad = []
    had = hadith[i-1]
    labels = ibn_majah_predictions[i].split(',')
    for count in range(len(labels)):
        label = labels[count]
        if label ==" 'B-PERS'":
            narrator.append(had[count])
        elif label ==" 'I-PERS'":
            narrator.append(had[count])
        else:
            if len(narrator) != 0:
                name = ' '.join(narrator)
                isnad.append(name)
            narrator = []
        count += 1
    data.append(isnad)

```

```python
narrators = pd.DataFrame(data=[data])
narrators = narrators.T
narrators.columns = ['Narrators']
narrators
```

```python
df = pd.concat([df, narrators],axis=1)
```
