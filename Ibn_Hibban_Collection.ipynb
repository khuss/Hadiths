```python
test = [24250,24245,24261,24217,24222,24237,24327,24357,24363]
df_trial = df_IbnHibban.loc[:,'clean_texts']
df_test = df.iloc[test].loc[:,'clean_texts']
```

```python
def is_numb(word):
    for w in word:
        if w  in ar_num:
            isit = True
        else:
            isit = False
            break
    return isit


        
def translate(num):
    try:
        eng_equi=''
        for i in range(0,len(num)):
                temp = ar_num.index(num[i])
                eng_equi += eng_num[temp]
    except:
        temp = 0
    return eng_equi


def multiplicity(worded_hadith):
    ind = []
    for i in range(len(worded_hadith)-1):
        b1 = is_numb(worded_hadith[i])  
        b2 = (worded_hadith[i+1])
        if (b1 and (b2 == '-')):
            ind.append(i)
    return ind

    
hadith = []  
index = []
hadith_number = []
ar_num = ['٨','٦','٩','٥','٣','٧','٢','٤','١',' ','-','(', ')','م', '٠','ب']
eng_num = ['8', '6', '9', '5', '3', '7', '2', '4','1', ' ', '-', '(', ')', 'M','0','B']
count = 0
for i in df_trial:    
    j = " ".join(i)
    w = j.split()
    ind = multiplicity(w)
    if len(w)>15:
        if (len(ind) != 0):
            for k in range(len(ind)):
                if k!= len(ind)-1:
                    had = w[ind[k]:ind[k+1]]
                    hadith.append(" ".join(had))
                    hadith_number.append(translate(w[ind[k]]))
                    index.append(df_trial.index[count])
                else:
                    hadith.append(" ".join(w[ind[k]:len(w)]))
                    hadith_number.append(translate(w[ind[k]]))
                    index.append(df_trial.index[count])  
        else:
            hadith.append(" ".join(w))
            hadith_number.append(0)
            index.append(df_trial.index[count])
    
    count+=1
    
    
```

```python
new_df = pd.DataFrame(hadith,index = index,columns=['clean_texts'])
hadith_number = pd.DataFrame(hadith_number, columns = ['hadith_number'])
```

```python
df_IbnHibban.drop(columns='clean_texts',inplace=True)
new_df = new_df.join(df_IbnHibban,how='left')
new_df = new_df.reset_index()
```

```python
new_df = pd.concat([new_df, hadith_number],axis=1)
```
